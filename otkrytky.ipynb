```python
# Импорт графических модулей
from PIL import Image, ImageDraw, ImageFont
import re

# Функция valinput запрашивает ввод данных с определенными вариантами, пока не будет введено корректное значение.
def valinput(prompt, validopts):
    uinput = input(prompt)
    while uinput not in validopts:
        print("Некорректный ввод. Попробуй еще раз. Выбери вариант из предложенных.")
        uinput = input()
    return uinput

#Функция valdate запрашивает ввод даты в формате дд.мм.гг., пока не будет введена корректная дата.
def valdate(prompt):
    dpattern = re.compile(r'^\d{2}\.\d{2}\.\d{2}$')
    uinput = input(prompt)
    while not dpattern.match(uinput):
        print("Некорректный ввод. Попробуй еще раз. Дата должна быть в формате дд.мм.гг.")
        uinput = input()
    return uinput

# выбор фона
backgrounds = ['голубой', 'зеленый', 'оранжевый', 'розовый', '']
backgroundp = 'Выбери фон для открытки. У нас есть: голубой / зеленый / оранжевый / розовый. \nВ строку напиши, например, "голубой": '
background = valinput(backgroundp, backgrounds) #выводит индекс введенной переменной backgroundp в списке backgrounds, чтобы после найти по индексу ключа найти значение в словаре color
color = {
    'голубой': (93, 180, 244),
    'зеленый': (34, 304, 123),
    'оранжевый': (255, 104, 2),
    'розовый': (255, 103, 103),
}.get(background, (255, 255, 255))

image = Image.new("RGB", (1080, 1080), color)
draw = ImageDraw.Draw(image)

# выбор шрифта
fonts = ['добрый', 'романтичный', 'веселый', 'строгий', 'детский', 'безумный']
fontp = 'Выбери характер шрифта: добрый / романтичный / веселый / строгий / детский / безумный\nВ строку напиши, например, "веселый": '
fontchoose = valinput(fontp, fonts) #выводит индекс введенной переменной fontp в списке fonts, чтобы после найти по индексу ключа найти значение в словаре font_name

# присваиваем каждой характеристики свою переменную со шрифтовым файлом
font_name = {
    'веселый': 'Shnobel-Oblique.ttf',
    'детский': 'TransformaScript_Trial-Regular.ttf',
    'безумный': 'YUNGA-Display.ttf',
    'добрый': 'overdozesans.ttf',
    'романтичный': 'alexandrascript.ttf',
    'строгий': 'bouquet.otf',
}[fontchoose]

# выбор картинки
pics = ['веселый кот', 'бабочка', 'хомяк', 'ежик', 'строгий кот', 'романтичная собака', 'кролик']
picp = 'Выбери украшения для открытки. У нас есть: веселый кот / бабочка / хомяк / ежик / строгий кот / романтичная собака / кролик\nВ строку напиши название картинки: '
pic = valinput(picp, pics) #выводит индекс введенной переменной picp в списке pics, чтобы после найти по индексу ключа найти значение в словаре pic_list
pic_list = {
    'веселый кот': 'cat.png',
    'бабочка': 'babochka.png',
    'хомяк': 'homik.png',
    'ежик': 'eshik.png',
    'строгий кот': 'cat_strogi.png',
    'романтичная собака': 'dog_romantic.png',
    'кролик': 'crolik.png'
}
picture = Image.open(pic_list[pic])

# выбор украшения
pic2o = ['конфетти', 'ромашки', 'золото', 'розы', 'снег']
pic2p = 'Выбери картинку для открытки. У нас есть: конфетти / ромашки / золото / розы / снег\nВ строку напиши название картинки: '
pic2 = valinput(pic2p, pic2o) #выводит индекс введенной переменной pic2p в списке pic2o, чтобы после найти по индексу ключа найти значение в словаре pic2_list
pic2_list = {
    'конфетти': 'konfeti.png',
    'ромашки': 'romashka.png',
    'золото': 'zoloto.png',
    'розы': 'rozi.png',
    'снег': 'snow.png'
}
picture2 = Image.open(pic2_list[pic2])

# добавление градиента
gradient = Image.open('gradient.png')
image.paste(gradient, (0, 0), mask=gradient)

# добавление украшения
image.paste(picture2, (300, 0), mask=picture2)

# добавление картинки
image.paste(picture, (0, 0), mask=picture)

# сбор текстовой информации
name_to = input('Кого ты хочешь поздравить? Введи имя в именительном падеже: ')
font = ImageFont.truetype(font_name, size=128)
# измеряем размер текстового блока, учитывая шрифт, его размер и добавление восклицательного знака
text_bbox = draw.textbbox((0, 0), name_to + '!', font=font)
# Вычисляем ширину и высоту текстового блока, вычитая координаты левого верхнего угла из координат правого нижнего угла.
# Это поможет нам получить размеры прямоугольной области, в которой будет отображен текст.
text_width, text_height = text_bbox[2] - text_bbox[0], text_bbox[3] - text_bbox[1]
x = (1080 - text_width) // 2
draw.text((x, 48), name_to + '!', font=font, fill=(255, 255, 255))

# повторяем те же действия для написания поздравления, меняем только размер шрифта
congrat_text = input('Напиши небольшое поздравление: ')
font = ImageFont.truetype(font_name, size=64)
text_bbox = draw.textbbox((0, 0), congrat_text, font=font)
text_width, text_height = text_bbox[2] - text_bbox[0], text_bbox[3] - text_bbox[1]
x = (1080 - text_width) // 2
draw.text((x, 210), congrat_text, font=font, fill=(255, 255, 255))

#используем встроенное в графическую библиотеку выравнивание по правой стороне и якорную точку для начала написания -- лево 
nfrom = input('Напиши своё имя для подписи открытки: ')
font = ImageFont.truetype(font_name, size=64)
draw.text((1050, 966), nfrom, font=font, fill=(255, 255, 255), align='left', anchor='rs')

day = valdate('Когда состоится праздник? Напиши дату в формате дд.мм.гг: ')
font = ImageFont.truetype(font_name, size=64)
draw.text((1050, 1047), day, font=font, fill=(255, 255, 255), align='left', anchor='rs')

pic2 = Image.open(pic2_list[pic2])

# выводим открытку
image.show()
image.save('открытка.jpeg')

print('--------------------------------------------------------------')
print("Открытка успешно сохранена в папку, где находится этот код!")
```
